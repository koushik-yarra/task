# Generating random data


```python
# Generating random data
import pandas as pd
import random

# Parameters
num_rows = 10000

# Data Pools
locations = ["Ahmedabad", "Surat", "Mumbai", "Delhi", "Pune", "Bangalore", "Chennai", "Kolkata","Hyderabad","Vizag"]
colleges = ["KLU Institute", "IIT Khargarpur", "IIT Bombay", "IIT Madras", "Pragati College", " UPGRAD academy", "JNTU","Analytics Vidhya"]
years_of_study = ["1st", "2nd", "3rd", "4th"]
program_interests = ["Data Science", "Robotics", "AI", "Electric Vehicle", "Web Development", "Full-stack", "Cybersecurity", "Devops"]
lead_sources = ["Instagram", "LinkedIn", "College Collaboration", "Google Form", "Mass-Mailing", "Whatsapp"]

# Data Generation
data = {
    "Lead ID": [f"LD{i+1}" for i in range(num_rows)],
    "Location": [random.choice(locations) for _ in range(num_rows)],
    "College": [random.choice(colleges) for _ in range(num_rows)],
    "Year of Study": [random.choice(years_of_study) for _ in range(num_rows)],
    "Program Interest": [random.choice(program_interests) for _ in range(num_rows)],
    "Lead Source": [random.choice(lead_sources) for _ in range(num_rows)],
}

# Create DataFrame
df = pd.DataFrame(data)

# Save the dataset to a CSV file
df.to_csv("lead.csv", index=False)

print("Dataset generated and saved as 'lead.csv'.")
```

# Demographic Analysis

```python
import matplotlib.pyplot as plt
```

```python
location_counts = df["Location"].value_counts()
```

```python
location_counts
```

```python
college_counts = df["College"].value_counts()
college_counts
```

```python
year_counts = df["Year of Study"].value_counts()
year_counts
```

```python
program_counts = df["Program Interest"].value_counts()
program_counts
```

# Visualization

```python
plt.figure(figsize=(15, 10))

# Location Analysis
plt.subplot(2, 2, 1)
location_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Lead Count by Location")
plt.xlabel("Location")
plt.ylabel("Count")
plt.xticks(rotation=45)

# College Analysis
plt.subplot(2, 2, 2)
college_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title("Lead Count by College")
plt.xlabel("College")
plt.ylabel("Count")
plt.xticks(rotation=45)

# Year of Study Analysis
plt.subplot(2, 2, 3)
year_counts.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightblue', 'lightgreen'])
plt.title("Lead Distribution by Year of Study")
plt.ylabel("")

# Program Interest Analysis
plt.subplot(2, 2, 4)
program_counts.plot(kind='bar', color='orange', edgecolor='black')
plt.title("Interest Levels for Programs")
plt.xlabel("Program")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Summary of Findings
summary = {
    "Top Location": location_counts.idxmax(),
    "Top College": college_counts.idxmax(),
    "Top Year of Study": year_counts.idxmax(),
    "Most Popular Program": program_counts.idxmax()
}

summary
```

```python
# Lead Source Analysis: Distribution of lead sources
lead_source_counts = df["Lead Source"].value_counts()
lead_source_counts
```

```python
# Program Interest by Year of Study
program_by_year = df.groupby(["Year of Study", "Program Interest"]).size().unstack(fill_value=0)
print(program_by_year)
```

```python
# Lead Source by Location
lead_source_by_location = df.groupby(["Location", "Lead Source"]).size().unstack(fill_value=0)
print(lead_source_by_location)
```

```python
# Visualizations
plt.figure(figsize=(18, 12))

# Lead Source Analysis
plt.subplot(2, 2, 1)
lead_source_counts.plot(kind='bar', color='purple', edgecolor='black')
plt.title("Lead Count by Source")
plt.xlabel("Lead Source")
plt.ylabel("Count")
plt.xticks(rotation=45)
```

```python
# Program Interest by Year of Study
plt.subplot(2, 2, 2)
program_by_year.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 5), ax=plt.gca())
plt.title("Program Interest by Year of Study")
plt.xlabel("Year of Study")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Program Interest")
```

```python
# Lead Source by Location (Heatmap Style)
plt.figure(figsize=(10, 8))
plt.imshow(lead_source_by_location, cmap='Blues', aspect='auto')
plt.title("Lead Source by Location")
plt.xlabel("Lead Source")
plt.ylabel("Location")
plt.xticks(range(len(lead_source_by_location.columns)), lead_source_by_location.columns, rotation=45)
plt.yticks(range(len(lead_source_by_location.index)), lead_source_by_location.index)
plt.colorbar(label="Count")
plt.tight_layout()
plt.show()
```

```python
# Findings from additional analysis
additional_summary = {
    "Top Lead Source": lead_source_counts.idxmax(),
    "Program Interest by Year": program_by_year.idxmax(axis=1).to_dict(),
    "Lead Source by Location (Max per Location)": lead_source_by_location.idxmax(axis=1).to_dict()
}

additional_summary
```
